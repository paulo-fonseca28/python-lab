```python
# MODULE 1 - NOTEBOOK

#Converting Celsius to Fahrenheit; Formula: F = (C * 9/5) + 32

# Lists from Step 1 (do not modify)
celsius_temperatures = [0, 10, 25, 32, 100]
fahrenheit_temperatures = [] 

# Convert each Celsius temperature to Fahrenheit
for celsius in celsius_temperatures:  # Start the loop (do not modify this line)
  fahrenheit = (celsius * 9/5) + 32

  fahrenheit_temperatures.append(fahrenheit)  # Append to the list (do not modify this line)

# Print the results (including the output from Step 1 - do not modify)
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)
```

```python
# Calculate total cost of items in a supermarket

supermarket_items = {"apples": 2.5, "bananas": 1.2, "oranges": 3.0, "grapes": 4.0}
total_cost = 0

for item, price in supermarket_items.items():
    total_cost += price
print(f"Total cost of all items: ${total_cost:.2f}")
```

```python
#Use a set to find unique numbers in a list

list_of_numbers = [1, 2, 3, 4, 5, 5, 5, 6, 6, 7, 8, 9, 9, 10]

unique_numbers = set(list_of_numbers)
print("Unique numbers:", unique_numbers)
```

```python
#Manual intersection of two lists

list_a = [1, 2, 3, 4, 5]
list_b = [4, 5, 6, 7, 8]
intersection = []

for item in list_a:
    if item in list_b:
        intersection.append(item)

print("Intersection:", intersection)
```
